Which albums stood out (AS IN THE BIGGEST)in terms of sales?

```python
import pandas as pd
from scipy.stats import zscore 
```

```python
pd.options.display.float_format = '{:,.2f}'.format

```

```python
lps = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
lps
```

```python
lps["sales_zscore"] = zscore(lps["sales"])
lps
```

```python
lps.head()
```

```python
dfs = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
dfs
```

```python
dfs["sales_zscore"] = zscore(dfs["streams_2023"])
dfs
```

sales=pd.concat([dfs,lps])
